## 1: bin <-> dec <-> hex

```python
def bin_dec(bin):
    dec = 0
    for i in range(len(str(bin))):
        dec += 2 ** (len(str(bin)) - i-1)*int(str(bin)[i])
    return(dec)
    
    
def dec_bin(dec):
    bin_ = ''
    for i in range (dec):
        if dec == 2 ** i:
            bin_ += '1' + i * '0'
            return bin_
    while dec != 0:
        bin_ += str(int(dec % 2))
        dec = int(dec / 2)
    return bin_[::-1]



def dec_hex(dec):
    hex_ = ''
    for i in range (dec):
        if dec == 16 ** i:
            hex_ += '1' + i * '0'
            return hex_
    while dec != 0:
        alist = ['1','2','3','4','5','6','7','8','9','A','B','C','D','E','F']
        hex_ += alist[int(dec % 16) - 1]
        dec = int(dec / 16)
    return hex_[::-1]

def hex_dec(hex):
    alist = ['1','2','3','4','5','6','7','8','9','A','B','C','D','E','F']
    blist = [alist.index(i)+1 for i in hex if i in alist]
    dec = 0
    for i in range(len(hex)):
        dec += blist[i] * 16 **(len(hex) - i -1)
    return(dec)
    
    

print(dec_bin(12331),
bin_dec(dec_bin(12331)),
dec_hex(2131),
hex_dec(dec_hex(2131)))
```

#### 2: convert 32bit word in floating point


```python
def float_word(number):
    if number > 0:
        sign = '0'
    else:
        sign = '1'
    mant = number - int(number)
    number = int(number-mant)
    number = str(dec_bin(number))
    def mant_bin(mant):
        bin_ = ''
        for i in range(23):
            if mant * 2 < 1:
                bin_ += '0'
                mant = mant * 2
            else:
                bin_ += '1'
                mant = 2 * mant - 1
        return(bin_)
    mant = mant_bin(mant)
    exp = len(number[1:])
    mant = number[1:] + mant
    mant = mant[:23]
    number = str(dec_bin(127 + exp))
    if len(number) != 8:
        number = (8 - len(number)) * '0' + number
    return (sign+number+mant)


# realized we have to do it the other way around

def word_float(word):
    word = str(word)
    sign = 1.0
    if word[0] == '1':
        sign = -1.0
    exp = 2 ** (bin_dec(int(word[1:9])) - 127)
    mant_bin = word[9:]
    mant_dec = 1.0
    for i in range(len(mant_bin)):
        mant_dec += int(mant_bin[i]) * 2 **(- i - 1)
    return (sign * exp * mant_dec)

print(float_word(2.22333333))
print(word_float(float_word(2.2233333)))
```

# 5. quadratic formular:

```python
from math import sqrt
def quad_form(a,b,c):
    x_plus = (-b + sqrt(b **2 - 4 * a * c))/(2 * a)
    x_minus = (-b - sqrt(b **2 - 4 * a * c))/(2 * a)
    return x_plus, x_minus

def quad_form2(a,b,c):
    x_plus = (b **2 - (b **2 - 4 * a * c))/((-2 * a * b) - 2 * a * sqrt(b **2 - 4 * a * c))
    x_minus = (b **2 - (b **2 - 4 * a * c))/((-2 * a * b) + 2 * a * sqrt(b **2 - 4 * a * c))
    return x_plus, x_minus

```

```python
def quad_form3(a,b,c):
    x_plus = (2 * c)/(-b - sqrt(b ** 2 - 4 * a * c))
    x_minus = (2 * c)/(- b + sqrt(b ** 2 - 4 * a * c))
    return x_plus, x_minus

print(quad_form(0.001, 1000, 0.001))
print(quad_form2(0.001, 1000, 0.001))
print(quad_form3(0.001, 1000, 0.001))
```

# 7. consider the integral of semicircle of radius 1:

```python
from math import sqrt
def semic(x):
    return sqrt(1 - x **2)
def integrate(func ,min_ , max_):
    area = 0
    N = 2000000
    for i in range(N):
        mid = abs((min_ - max_)) * 0.5 
        area +=  func(min_ + (i + 1) * mid/(N + 1)) * 2/N
    return area

print(integrate(semic, -1, 1))    
#%timeit integrate(semic, -1, 1)
```
